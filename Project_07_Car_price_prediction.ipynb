**Importing libraries:**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

**Upload the data file:**

**Distribution of catagorical data:**

```python
print(car_data.fuel.value_counts())
print(car_data.seller_type.value_counts())
print(car_data.transmission.value_counts())
```

**Performing label encoding:**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
car_data['fuel_encoded'] = label.fit_transform(car_data['fuel'])
car_data['seller_type_encoded'] = label.fit_transform(car_data['seller_type'])
car_data['transmission_encoded'] = label.fit_transform(car_data['transmission'])
car_data['owner_encoded']=label.fit_transform(car_data['owner'])
```

```python
car_data.head(3)
```

```python
new_car_data = car_data.drop(['fuel','seller_type','transmission','name','owner'],axis=1)
new_car_data.head(3)
```

```python
new_car_data.shape
```

**Splitting the data set:**

```python
X = new_car_data.drop(['selling_price'],axis=1)
Y = new_car_data['selling_price']
```

 **Train test splitting**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

**Linear regression**

```python
le_reg= LinearRegression()
```

```python
le_reg.fit(X_train,Y_train)
```

```python
train = le_reg.predict(X_train)
```

```python
error=metrics.r2_score(Y_train,train)
error
```

```python
test = le_reg.predict(X_test)
```

```python
test_error=metrics.r2_score(Y_test,test)
test_error
```
